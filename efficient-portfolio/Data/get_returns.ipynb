```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
file_name = 'stock_returns.csv'
lista = os.listdir()[1:]
lista.remove('get_returns.ipynb')
if file_name in lista:
    lista.remove(file_name)
```

```python
stock_returns = []
i = 0
for csv_file in lista:
    file = pd.read_csv(csv_file)
    returns = [100*(file['Adj Close'][i]/file['Adj Close'][i+1] -1) for i in range(file.Close.shape[0]-1)]
    stock_returns.append(returns) 
    i+=1
```

```python
with open(file_name, 'w') as new_file:    
    for j in range(len(stock_returns[0])):
        for i in range(len(stock_returns)):
            if i < len(stock_returns)-1: new_file.write(str(round(stock_returns[i][j],4))+';')
            else: new_file.write(str(round(stock_returns[i][j],4)))
        new_file.write('\n')
```
